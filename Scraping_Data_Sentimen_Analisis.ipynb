2. Import Library

```python
import pandas as pd
from google_play_scraper import reviews, Sort
```

3. Menentukan Aplikasi

```python
app_id = "com.ocbcnisp.onemobileapp"
```

4. Melakukan Scraping

```python
result, continuation_token = reviews(
    app_id,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=15000
)
```

5. Cek Jumlah Review

```python
print("Jumlah Review :", len(result))
```

6. Membuat DataFrame

```python
data = []

for review in result:

    data.append({

        'Username': review['userName'],
        'Review': review['content'],
        'Rating': review['score'],
        'Tanggal': review['at']

    })

df = pd.DataFrame(data)
```

11. Menghapus Data Kosong

```python
df = df.dropna()
```

12. Menghapus Review Duplikat

```python
df = df.drop_duplicates(subset=['Review'])
```

13. Simpan Dataset

```python
df.to_csv(
    "hasil_scraping_OCBC.csv",
    index=False,
    encoding="utf-8-sig"
)
```
