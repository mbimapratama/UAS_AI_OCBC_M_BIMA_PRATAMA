2 — Import Library

```python
import re
import string
import joblib
import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

3 — Load Model

```python
model = joblib.load("svm_model.pkl")

tfidf = joblib.load("tfidf_vectorizer.pkl")
```

4 — Stopword & Stemmer

```python
stop_words = set(stopwords.words("indonesian"))

factory = StemmerFactory()

stemmer = factory.create_stemmer()
```

5 — Fungsi Preprocessing

```python
def preprocessing(text):

    # lowercase
    text = text.lower()

    # hapus url
    text = re.sub(r"http\S+", "", text)

    # hapus mention
    text = re.sub(r"@\w+", "", text)

    # hapus hashtag
    text = re.sub(r"#\w+", "", text)

    # hapus angka
    text = re.sub(r"\d+", "", text)

    # hapus tanda baca
    text = re.sub(r"[^\w\s]", "", text)

    # tokenizing
    tokens = word_tokenize(text)

    # stopword
    tokens = [word for word in tokens if word not in stop_words]

    # stemming
    tokens = [stemmer.stem(word) for word in tokens]

    return " ".join(tokens)
```

6 — Prediksi Satu Kalimat

```python
kalimat = "Aplikasi sangat bagus dan mudah digunakan"

hasil = preprocessing(kalimat)

vector = tfidf.transform([hasil])

prediksi = model.predict(vector)

print("Kalimat :", kalimat)

print("Hasil :", prediksi[0])
```

7 — Coba Kalimat Negatif

```python
kalimat = "Login selalu gagal setelah update"

hasil = preprocessing(kalimat)

vector = tfidf.transform([hasil])

prediksi = model.predict(vector)

print(prediksi[0])
```

8 — Coba Kalimat Netral

```python
kalimat = "Aplikasi cukup baik namun masih perlu perbaikan"

hasil = preprocessing(kalimat)

vector = tfidf.transform([hasil])

prediksi = model.predict(vector)

print(prediksi[0])
```

(Jika model tidak pernah melihat cukup banyak data netral saat pelatihan, hasil bisa saja menjadi Positif atau Negatif. Itu hal yang wajar.)

9 — Prediksi Banyak Kalimat Sekaligus

```python
data = [

    "Transfer sangat cepat",

    "Aplikasi sering error",

    "Login mudah",

    "Kecewa karena selalu force close",

    "Pelayanannya sangat memuaskan"

]

for kalimat in data:

    hasil = preprocessing(kalimat)

    vector = tfidf.transform([hasil])

    prediksi = model.predict(vector)

    print("--------------------------------")

    print("Kalimat :", kalimat)

    print("Sentimen :", prediksi[0])
```

10 — Fungsi Interaktif

```python
def prediksi_sentimen():

    kalimat = input("Masukkan Kalimat : ")

    hasil = preprocessing(kalimat)

    vector = tfidf.transform([hasil])

    prediksi = model.predict(vector)

    print()

    print("Kalimat :", kalimat)

    print("Sentimen :", prediksi[0])
```
